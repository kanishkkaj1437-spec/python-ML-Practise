```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#supress warning for clean notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("Melbourne_housing_FULL.csv")
df.head()
```

```python
df.nunique()
```

```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
df = df[cols_to_use]
```

```python
df.head()
```

```python
df.shape
#rows,columns
```

```python
df.isna().sum()
```

```python
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
df[cols_to_fill_zero] = df[cols_to_fill_zero].fillna(0)
```

```python
df['Landsize'] = df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea'] = df['BuildingArea'].fillna(df.BuildingArea.mean())
```

```python
df.isna().sum()
```

```python
#Dropping NA columns
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le_suburb =  LabelEncoder()
le_type =  LabelEncoder()
le_method =  LabelEncoder()
le_sellerg =  LabelEncoder()
le_region =  LabelEncoder()
le_council =  LabelEncoder()
```

```python
df['suburb_n'] = le_suburb.fit_transform(df['Suburb'])
```

```python
df['type_n'] = le_type.fit_transform(df['Type'])
```

```python
df['meathod_n'] = le_method.fit_transform(df['Method'])
```

```python
df['seller_g'] = le_sellerg.fit_transform(df['SellerG'])
```

```python
df['region'] =  le_region.fit_transform(df['Regionname'])
```

```python
df['council'] = le_council.fit_transform(df['CouncilArea'])
```

```python
df.head()
```

```python
df = df.drop(['Suburb','Type','Method','SellerG','Regionname','CouncilArea'],axis = 'columns')
```

```python
df.head()
```

```python
X = df.drop('Price', axis=1)
y = df['Price']
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```

```python
reg.score(X_train,y_train)
```

```python
#X_train
```

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_reg.fit(X_train, y_train)
```

```python
lasso_reg.score(X_test,y_test)
```

```python
from sklearn.linear_model import Ridge
ridge_reg= Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_reg.fit(X_train,y_train)
```

```python
ridge_reg.score(X_test,y_test)
```
