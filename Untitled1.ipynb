```python
import math
import time as time
a = 141421
# a = 1413
f = 4 * pow(10, 20)
dr = 1
suuy = 0
maxx = 1
tic = time.time()
for i in range(2, a+1, 2):
    za = f - (i*i)
    zb = 1 + (i*i)
    zc = za/zb
    b = int(math.sqrt(zc))
    if i == 2:
        xd = time.time()
        print(1000*(xd-tic), i, b)
    else:
        xf = time.time()
        print(1000*(xf-xd), i)
        xd = xf

    for j in range(i+2, b+1, 2):
        summ = 0
        aa = i*i
        bb = j*j
        cc = (i*j)*(i*j)
        summ = aa + bb + cc

        dd = int(math.sqrt(summ))

        if summ == dd**2:
            if dd % 2 == 0:
                area = int(dd/2)
                xa = pow(10, 10)
                if area < xa:
                    if area > maxx:
                        maxx = area
                    suuy = suuy + area
                    print(i, j, summ, dd, area, dr)
                    dr = dr + 1

print("hasdgjhwafb")
print(maxx)
print(suuy)
```
