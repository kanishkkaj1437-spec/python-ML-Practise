```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1,2,3,4]
y = [2,4,6,8]
#resize of figure
plt.figure(figsize=(5,3),dpi=300)

plt.plot(x,y,label='2x',color="red",linewidth="2",marker=".",linestyle="--",markersize=10,markeredgecolor="blue")

x2 = np.arange(0,4.5,0.5)
plt.plot(x2,x2**2,'b',label="x**2")
plt.title("Graph",fontdict={"fontname":"Comic Sans MS","fontsize":20})
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.xticks([1,2,3,4,10])
plt.yticks([2,3,4,6,8,10])

#save graph
plt.savefig('my_graph.png',dpi=300)

plt.legend()
plt.show()
```

### Bar Chart

```python
labels = ['A','B','C']
values = [4,10,15]

bars = plt.bar(labels,values)

bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')

plt.figure(figsize=(6,4))

plt.show()
```

# Real World Examples

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(6,4),dpi=200)
plt.title("Gas Prices over time (in USD)")




#plt.plot(gas.Year,gas.USA,'b.-',label='USA')
#plt.plot(gas["Year"],gas['Canada'],'r.-',label='Canada')
#plt.plot(gas.Year,gas["South Korea"],'g.-',label='South Korea')

for country in gas:
    if country != "Year":
        plt.plot(gas.Year,gas[country],'.-',label=country)
        
        



plt.xticks(gas.Year[::3])

plt.xlabel("Year")
plt.ylabel("Prices (in USD)")
plt.legend()
plt.show()
```

# Second Example



```python
fifa = pd.read_csv("fifa_data.csv")
fifa.head()
```

# Histograms

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins = bins)

plt.xticks(bins)

plt.title("Distribution of Player skill in fifa 2018")
plt.ylabel("No of Players")
plt.xlabel("Overall Skill")

plt.show()
```

# Pie Chart

```python
data=fifa[fifa["Preferred Foot"]=="Left"]
left = data.count()[0]
right = fifa.loc[fifa["Preferred Foot"] =="Right"].count()[0]

labels = ['Left','Right']
plt.pie([left,right],labels=labels,autopct = '%.2f %%')

plt.title("Foot Preferrence of FIFA Players")
plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

plt.style.use("ggplot")

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight <150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight > 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight > 200].count()[0]

weights = [light,light_medium,medium,medium_heavy,heavy]
label = ['Under 125', '125-150', '150-175','175-200','Over 200']
explode = (0.4,0.2,0,0,0.4)

plt.title("Weight of Professional Soccer Players (lbs)")

plt.pie(weights,labels = label, explode = explode,autopct = '%.2f %%',pctdistance=0.8)

plt.show()
```

```python
plt.style.use('default')
plt.figure(figsize=(5,8),dpi=100)
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']["Overall"]
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['Barcelona','Real Madrid','NE Revolution']
bp=plt.boxplot([barcelona,madrid,revs],labels=labels, patch_artist = True,medianprops=dict(linewidth=2))

plt.title("Overall Comparison of top football clubs")
plt.ylabel('Overall Ratings')

for box in bp['boxes']:
    box.set(color='blue',linewidth=2)
    box.set(facecolor='#e0e0e0')
    #box.set(hatch = '/')

plt.show()
```
